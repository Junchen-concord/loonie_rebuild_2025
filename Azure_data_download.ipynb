```python
import joblib

# Paths
isgood_path = "/Users/starsrain/2025_concord/loonie_rebuild_2025/models_store/IsGood_V1_TestBed.pkl"
ibv_path    = "/Users/starsrain/2025_concord/loonie_rebuild_2025/models_store/IBV_V3_TestBed.pkl"

# Load models
model_isgood = joblib.load(isgood_path)
model_ibv    = joblib.load(ibv_path)

print("Models loaded successfully")
print("IsGood model type:", type(model_isgood))
print("IBV/FPDAA model type:", type(model_ibv))
```
